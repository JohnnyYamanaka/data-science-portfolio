A abordagem aqui é simples. Podemos ver que aparentemente os dados até 2000 são nulos, dessa forma eles serão eliminados.  
Mas em alguns anos vemos que existe o indicador para alguns valores e outros não, e nesse caso ainda eles não serão deletados.  
Assim sendo, será criado uma função para verificar se a linha toda contém nulo e somente nesse caso deletar a linha.

```python
def eliminate_null(df):
    df = df[df.sum(axis=1, numeric_only=True) > 0]
    df.reset_index(inplace=True, drop=True)
    
    return df
```

```python
for pais in lista_paises:
    exec(f'{pais} = eliminate_null({pais})')
```

## Verificando os indicadores  
Agora vamos passar indicador por indicador para verificar se existem outliers

```python
#Verificar se para o dataframe existe o indicador desejado
def is_column_exist(pais, indicador):
    return indicador in pais.columns

#Caso exista o indicador, imprimir a estatistica dele
def get_estatistica(pais, indicador):
    if is_column_exist(pais, indicador):
        print(pais[indicador].describe())
        print('\n')

    else:
        print('indicador nao existe \n')
```

A Argentina contem outliers nos anos de 2017 ate 2020, enquanto a Guiana Francesa tem um no anos 2000. Para preencher os valores, vou pegar os valores 'vizinhos' e tirar uma media.  
Pode nao ser a melhor abordagem, mas pela simplicidade vou fazer dessa forma.

```python
index_to_fix = [17, 18, 19, 20]

for index in index_to_fix:
    AR['cod_77832'].iloc[index] = AR['cod_77832'][14:17].mean()
    
AR['cod_77832']
```
