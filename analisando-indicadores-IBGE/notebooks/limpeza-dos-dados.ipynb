A abordagem aqui  simples. Podemos ver que aparentemente os dados at 2000 so nulos, dessa forma eles sero eliminados.  
Mas em alguns anos vemos que existe o indicador para alguns valores e outros no, e nesse caso ainda eles no sero deletados.  
Assim sendo, ser criado uma funo para verificar se a linha toda contm nulo e somente nesse caso deletar a linha.

```python
BR[BR.sum(axis=1, numeric_only=True) > 0]
```

```python
def eliminate_null(df):
    df = df[df.sum(axis=1, numeric_only=True) > 0]
    df.reset_index(inplace=True, drop=True)
    
    return df
```

```python
for pais in lista_paises:
    exec(f'{pais} = eliminate_null({pais})')
```

## Verificando os indicadores  
Agora vamos passar indicador por indicador para verificar se existem outliers

```python
#Verificar se para o dataframe existe o indicador desejado
def is_column_exist(pais, indicador):
    return indicador in pais.columns

#Caso exista o indicador, imprimir a estatistica dele
def get_estatistica(pais, indicador):
    if is_column_exist(pais, indicador):
        print(pais[indicador].describe())
        print('\n')

    else:
        print('indicador nao existe \n')
```

### Cod_77821 (Investimentos em pesquisa e desenvolvimento - % do PIB)
 possvel notar que, em alguns os dataframes, o ano de 2018 est com valor errado - o valor  sempre 99999999999998%. Dessa forma como estamos tratando de intervalo de valor muito grande o pandas mostrar os valores em notao cientfica.  
J que esse valor est errado, esse valor ser substiudo por 0.9988, o que representa 99,99%.

```python
def verificar_outlier_77821(df):
    if ('cod_77821' in df.columns):
        print(df['cod_77821'].iloc[18])
    else:
        pass
```

```python
for pais in lista_paises:
    exec(f"print('{pais}: '), verificar_outlier_77821({pais})")
#     exec(f'print("{pais}: verificar_outlier_77821({pais})")')
```

A Guiana Francesa (GF) nao possui o indicador 77821, e a Colombia (CO), Peru (PE) parecem estar com os valores corretos, entao a alteracao nao sera aplicada a eles.

```python
paises_para_tratar_77821 = [pais for pais in lista_paises if pais not in ['GF', 'CO', 'PE']]
paises_para_tratar_77821
```

```python
for pais in paises_para_tratar_77821:
    exec(f"{pais}['cod_77821'].iloc[18] = .9999")
    exec(f"print('{pais}: ', {pais}['cod_77821'].iloc[18])")
```

A Argentina contem outliers nos anos de 2017 ate 2020, enquanto a Guiana Francesa tem um no anos 2000. Para preencher os valores, vou pegar os valores 'vizinhos' e tirar uma media.  
Pode nao ser a melhor abordagem, mas pela simplicidade vou fazer dessa forma.
